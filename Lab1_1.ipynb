7. Рассмотрим случайную матрицу  $15 × 2 $, представляющую декартовы координаты, преобразуйте их в полярные координаты.

```python
m = np.random.rand(15, 2)
x = m[:, 0]
y = m[:, 1]

r = np.sqrt(x**2 + y**2)
ang = np.arctan2(y, x)

m = np.column_stack((r, ang))
m
```

8. По двум массивам $x$ и $y$ постройте матрицу Коши $C$
 $(C_{ij} =\frac{1}{x_i - y_j})$

```python
x = np.random.rand(5)
y = np.random.rand(5)

x = x.reshape(-1, 1)

diff = x - y
C = 1 / diff
C
```

9. Рассмотрим матрицу $10 × 3 $, извлеките строки с неравными значениями (например, $[2,2,3]$).

# 2. Pandas

# Найдите ответы на вопросы с помощью предложенного датасета:

Файл с датасетом - `olympics_dataset.csv`

```python
import pandas as pd
```

```python
df = pd.read_csv('olympics_dataset.csv')
df
```

1. Сколько спортсменов участвовали в Летних Олимпийских играх за всё время?

```python
df[df['Season'] == 'Summer']['player_id'].nunique()
```

2. Найдите участника с наибольшим количеством медалей

```python
medal_winners = df[df['Medal'] != 'No medal']
top_athlete = medal_winners.groupby(['player_id', 'Name'])['Medal'].count().reset_index()
top_athlete.sort_values(by='Medal', ascending=False).iloc[0]['Name']
```

3. Какой спорт является самым популярным, то есть в нём участвовало больше всего спортсменов за всё время?

```python
popular_sport = df.groupby('Sport')['player_id'].nunique().reset_index()
popular_sport.sort_values(by='player_id', ascending=False).iloc[0]['Sport']
```

4. Кто был победителем медального зачёта в 2004 году?

```python
year_2004 = df[(df['Year'] == 2004) & (df['Medal'] != 'No medal')]
medal_counts = year_2004.groupby('Team')['Medal'].count().reset_index()
medal_counts.sort_values(by='Medal', ascending=False).iloc[0]['Team']
```

5. В каком виде спорта США завоевало наибольшее число медалей?

```python
usa_medals = df[(df['Team'] == 'United States') & (df['Medal'] != 'No medal')]
sport_medals = usa_medals.groupby('Sport')['Medal'].count().reset_index()
sport_medals.sort_values(by='Medal', ascending=False).iloc[0]['Sport']
```

# С помощью графика покажите:

6. Сколько женщин и мужчин участвовало в Летних Олимпийских играх каждый год?

```python
import matplotlib.pyplot as plt
summer_sportsmen = df[df['Season'] == 'Summer'].groupby(['Year', 'Sex'])['player_id'].nunique().unstack()
summer_sportsmen.plot(kind='bar', stacked=True)
plt.title('Участие мужчин и женщин в летних Олимпиадах')
plt.xlabel('Год')
plt.ylabel('Количество участников')
plt.show()
```

7. Какие 10 стран на текущий момент имеют наибольшее количество золотых медалей?


```python
gold_medals = df[df['Medal'] == 'Gold'].groupby('Team')['Medal'].count()
gold_medals.nlargest(10).sort_values().plot(kind='barh')
plt.title('Топ-10 стран по золотым медалям')
plt.xlabel('Количество золотых медалей')
plt.show()
```

8. Сколько золотых, серебряных и бронзовых медалей зарабатывал СССР отдельно по каждым играм?

```python
ussr_medals = df[(df['NOC'] == 'URS') & (df['Medal'] != 'No medal')]
ussr_pivot = ussr_medals.pivot_table(index='Year', columns='Medal', values='player_id', aggfunc='count', fill_value=0)
ussr_pivot.plot(kind='bar')
plt.title('Медали СССР по типам')
plt.xlabel('Год Олимпиады')
plt.ylabel('Количество медалей')
plt.show()
```

9. Сколько медалей завоёвывали спортсмены и спортсменки (отдельно) команды Китая на каждой олимпиаде?


```python
china_medals = df[(df['Team'] == 'China') & (df['Medal'] != 'No medal')]
china_gender = china_medals.groupby(['Year', 'Sex'])['Medal'].count().unstack()
china_gender.plot(marker='o')
plt.title('Медали Китая по полу')
plt.xlabel('Год Олимпиады')
plt.ylabel('Количество медалей')
plt.grid(True)
plt.show()
```

10. Какие 10 стран имеют наибольшее количество участников, не получивших никаких медалей ни на одной из Летних олимпиад?

```python
summer_non_medal = df[(df['Season'] == 'Summer') & (df['Medal'] == 'No medal')]
non_medal_count = summer_non_medal.groupby('Team')['player_id'].nunique().nlargest(10)
non_medal_count.sort_values().plot(kind='barh', color='grey')
plt.title('Страны с наибольшим числом участников без медалей')
plt.xlabel('Количество участников')
plt.show()
```
